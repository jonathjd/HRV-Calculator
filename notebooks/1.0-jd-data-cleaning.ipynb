```python
date_cols = ['creationDate', 'startDate', 'endDate']

for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

```python
df['type'] = df['type'].str.replace('HKQuantityTypeIdentifier', '')
df['type'] = df['type'].str.replace('HKCategoryTypeIdentifier', '')
df['workoutActivityType'] = df['workoutActivityType'].str.replace('HKWorkoutActivityType', '')
```

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['value'] = df['value'].fillna(1.0)

df['duration'] = pd.to_numeric(df['duration'])
df['activeEnergyBurned'] = pd.to_numeric(df['activeEnergyBurned'])
```

```python
df['MetadataEntry'].unique()
```

```python
df_cleaned = df.drop(columns=['HKCharacteristicTypeIdentifierFitzpatrickSkinType', 
                              'HeartRateVariabilityMetadataList', 
                              'HKCharacteristicTypeIdentifierCardioFitnessMedicationsUse',
                              'HKCharacteristicTypeIdentifierBloodType',
                              'HKCharacteristicTypeIdentifierBiologicalSex',
                              'HKCharacteristicTypeIdentifierDateOfBirth',
                              'WorkoutEvent',
                              'WorkoutRoute',
                              'WorkoutStatistics',
                              'MetadataEntry'
                              ]).copy()
```
